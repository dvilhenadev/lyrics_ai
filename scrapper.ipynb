```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
artistList = ["Samuel Úria", "Tiago Bettencourt", "Lena d'Água", "António Variações", "Rui Veloso", "João Só", "Xutos e Pontapés", "Os Quatro e Meia", "Miguel Araújo", "Os azeitonas"]
```

```python
song_dict = {
    "á":"a",
    "à":"a",
    "ã":"a",
    "â":"a",
    "é":"e",
    "ê":"e",
    "í":"i",
    "ô":"o",
    "ó":"o",
    "õ":"o",
    "ú":"u",
    "ç":"c",
    " ":"_",
    "-":"_",
    ",":"_",
    ".":"_",
    ":":"_",
    ";":"_"
}

artist_dict = {
    "á":"a",
    "à":"a",
    "ã":"a",
    "â":"a",
    "é":"e",
    "ê":"e",
    "í":"i",
    "ô":"o",
    "ó":"o",
    "õ":"o",
    "ú":"u",
    "ç":"c",
    " ":"-",
    "\'":""
}

htmlsong_dict = {
    "<br>":"\n",
    "</br>":"\n",
    "<br/>":"\n",
    "</p>":"\n\n",
    "<p>":"\n"
}
```

```python
for artist in artistList:

    artist = artist.lower()

    artist = ''.join(artist_dict.get(ch, ch) for ch in artist)

    websiteURL = "https://www.letras.com/"
    artistURL = websiteURL + artist.lower() + "/"

    content = urllib.request.urlopen(artistURL)
    read_content = content.read()
    soup = BeautifulSoup(read_content,'html.parser')

    fullDiv = soup.find("div", {"class": "artista-top g-sp g-pr"})
    listcontainer = fullDiv.find("div", {"class": "list-container"})
    listitem = listcontainer.find_all("li", {"class": "cnt-list-row -song"})

    fieldList = ["data-shareurl", "title"]


    for song in listitem:
    # You want to get: data-shareurl , data-url
    #print(song)
        songStr = str(song)
        dataList = []
    
        for field in fieldList:
            startIndex = songStr.find(field)
            quotationStartIndex = songStr.find("\"", startIndex)+1
            quotationEndIndex = songStr.find("\"", quotationStartIndex)
            dataStr = songStr[quotationStartIndex:quotationEndIndex]
            dataList.append(dataStr)
        #print(fieldStr)
        
        songURL = dataList[0]
    
        songContent = urllib.request.urlopen(songURL)
        read_songContent = songContent.read()
        song_soup = BeautifulSoup(read_songContent, 'html.parser')
    
        lyrics_div = song_soup.find("div", {"class": "cnt-letra p402_premium"})
        lyrics_pAll = lyrics_div.find_all("p")
    
        lyrics_string = ""
    
        for lyrics in lyrics_pAll:
            verse = str(lyrics)
        #print(verse)
            verse = verse.replace("<br>", "\n")
            verse = verse.replace("</br></p>", "\n\n")
            verse = verse.replace("</br>", "\n")
            verse = verse.replace("<br/>", "\n")
            verse = verse.replace("<p>", "")
            verse = verse.replace("</p>", "\n\n")
        
            lyrics_string += verse
    
        if lyrics_string.endswith("\n"):
            lyrics_string = lyrics_string[:-2]
        
        
  
        songFileName = ''.join(song_dict.get(ch, ch) for ch in dataList[1]).lower()+".txt"
        filePath = "dataset/"+songFileName
    
        with open(filePath, 'w',encoding="utf8") as f:
            f.write(lyrics_string)
```
